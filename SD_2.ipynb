```python
# Double Linked List

class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None
        self.prevval = None


class DoubleLinkedList:
    def __init__(self):
        self.headval = None

    # menampilkan isi list
    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval

    # tambah node di akhir
    def AddEnd(self, newdata):
        NewNode = Node(newdata)
        if self.headval is None:
            self.headval = NewNode
            return
        last = self.headval
        while last.nextval:
            last = last.nextval
        last.nextval = NewNode
        NewNode.prevval = last

    # sisipkan di tengah (setelah node tertentu)
    def AddMiddle(self, prev_node, newdata):
        if prev_node is None:
            print("Node tidak ditemukan")
            return

        NewNode = Node(newdata)
        NewNode.nextval = prev_node.nextval
        prev_node.nextval = NewNode
        NewNode.prevval = prev_node

        if NewNode.nextval is not None:
            NewNode.nextval.prevval = NewNode

    # hapus berdasarkan kata kunci
    def RemoveNode(self, key):
        temp = self.headval

        while temp is not None:
            if temp.dataval == key:
                # jika node pertama
                if temp.prevval is None:
                    self.headval = temp.nextval
                    if temp.nextval:
                        temp.nextval.prevval = None
                else:
                    temp.prevval.nextval = temp.nextval
                    if temp.nextval:
                        temp.nextval.prevval = temp.prevval
                return
            temp = temp.nextval

        print("Data tidak ditemukan")


# Program utama
dll = DoubleLinkedList()

# buat 3 node
dll.AddEnd("Durian")
dll.AddEnd("Mangga")
dll.AddEnd("Pisang")
print("Data awal:")
dll.listprint()

# sisipkan di tengah
dll.AddMiddle(dll.headval.nextval, "Alpukat")
print("\nSetelah penyisipan tengah:")
dll.listprint()

# hapus berdasarkan kata kunci
dll.RemoveNode("Mangga")
print("\nSetelah menghapus 'Mangga':")
dll.listprint()
```

```python
# Circular Linked List

class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None


class CircularLinkedList:
    def __init__(self):
        self.headval = None

    # menampilkan isi circular linked list
    def listprint(self):
        if self.headval is None:
            return

        printval = self.headval
        while True:
            print(printval.dataval)
            printval = printval.nextval
            if printval == self.headval:
                break

    # tambah node di akhir
    def AddEnd(self, newdata):
        NewNode = Node(newdata)

        if self.headval is None:
            self.headval = NewNode
            NewNode.nextval = self.headval
            return

        temp = self.headval
        while temp.nextval != self.headval:
            temp = temp.nextval

        temp.nextval = NewNode
        NewNode.nextval = self.headval

    # hapus node terakhir
    def RemoveEnd(self):
        if self.headval is None:
            return

        temp = self.headval

        # jika hanya satu node
        if temp.nextval == self.headval:
            self.headval = None
            return

        prev = None
        while temp.nextval != self.headval:
            prev = temp
            temp = temp.nextval

        prev.nextval = self.headval


# Program utama
cll = CircularLinkedList()

# buat 3 node
cll.AddEnd("Durian")
cll.AddEnd("Mangga")
cll.AddEnd("Pisang")

print("Data awal:")
cll.listprint()

# penyisipan di akhir
cll.AddEnd("Alpukat")

print("\nSetelah penyisipan akhir:")
cll.listprint()

# hapus node terakhir
cll.RemoveEnd()

print("\nSetelah penghapusan akhir:")
cll.listprint()
```
